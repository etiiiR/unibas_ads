```python
# binary search
def binary_search_lower_bound(data: list, value: int):
    l: int = -1
    h: int = len(data)
    while l + 1 < h:
        mid: int = (l + h) // 2
        if value > data[int(mid)]:
            l = mid
        else:
            h = mid
    return h


# binary search upper bound
def binary_search_upper_bound(data: list, value: int):
    l: int = -1
    h: int = len(data)
    while l + 1 < h:
        mid: int = (l + h) // 2
        if value >= data[int(mid)]:
            l = mid
        else:
            h = mid
    return l
```

### Sorting algos

#### Selection Sort

- General Idea first loop through whole list
- Second loop through elements where we swap the two values
- Sorted part is always at the left side so the left side is left (sorted) <= right side (unsorted)

```python
def selection_sort(data: list):
    n: int = len(data) - 1
    for i in range(n):
        # SEARCH: minimal element within the loop
        m = i
        # second loop remeber the minimal element change it if its lower
        for j in range(i + 1, len(data)):
            if data[j] < data[m]:
                m = j
        # swap the element pos: i, m: minindex,
        data[i], data[m] = data[m], data[i]
    return data
```

### Insertion Sort
general Idea: 

```python
def insertion_sort(data: list):
    n: int = len(data)
    for i in range(1, n):
        j = i
        while j>0 and data[j-1] > data[j]:
            data[j], data[j-1] = data[j-1], data[j]
            j = j - 1
    return data
            
```

```python
import dataclasses
from dataclasses import field


@dataclasses.dataclass(order=True)
class Person:
    name: str = field(compare=False)
    age: int

    def __str__(self) -> str:
        return f"{self.name} ({self.age})"
```

```python
beat = Person("Beat", 15)
hans = Person("Hans", 11)
gabi = Person("Gabi", 1)
```

```python
array_of_person = [beat]
array_of_person.append(hans)
array_of_person.append(gabi)
```

```python
selection_sort(array_of_person)
```
