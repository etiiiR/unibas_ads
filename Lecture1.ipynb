```python
# binary search
def binary_search_lower_bound(data: list, value: int):
    l:int = -1; h:int = len(data)
    while(l + 1 < h):
        mid:int = (l + h) // 2
        if (value > data[int(mid)]):
            l = mid
        else:
            h = mid
    return h


# binary search upper bound 
def binary_search_upper_bound(data: list, value: int):
    l:int = -1; h:int = len(data)
    while(l + 1 < h):
        mid:int = (l + h) // 2
        if (value >= data[int(mid)]):
            l = mid
        else:
            h = mid
    return l
```

```python
binary_search_lower_bound([1,2,3,4,5,6], 4)
```

### Sorting algos

#### Selection Sort

- General Idea first loop through whole list
- Second loop through elements where we swap the two values
- Sorted part is always at the left side so the left side is left (sorted) <= right side (unsorted)

```python
def selection_sort(data: list):
    n: int = len(data) -1
    for i in range(n):
        # SEARCH: minimal element within the loop
        m = i
        # second loop remeber the minimal element change it if its lower
        # then swape the element
        for j in range(i + 1, len(data)):
            if data[j] < data[m]:
                m = j
        data[i], data[m] = data[m], data[i]
    return data
```

```python
res = selection_sort([1,5,3,4,5,6,2])
display(res)
```
